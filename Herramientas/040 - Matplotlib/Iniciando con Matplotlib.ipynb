# Iniciando con Matplotlib

```python
import matplotlib.pyplot as plt

import numpy as np


```

```python
%matplotlib inline
x = np.linspace(0,10,100)

dfig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

```python
x= np.linspace(0,10,100)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label = 'Seno')#Linea  Punteada
ax.plot(x, np.cos(x), '-', label = 'Coseno');
ax.axis('equal')
leg = ax.legend(loc='upper right', frameon = False)
```

# Subplots

```python
# Dividir en varios marcos las Graficacion

fig = plt.figure()

ax1 = fig.add_axes([0.1,0.5,0.8,0.4],
                  xticklabels=[], ylim=(-1.2,1.2))
ax2 = fig.add_axes([0.1,0.1,0.8,0.4],
                  ylim=(-1.2,1.2))

x= np.linspace(0,10)

ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```
